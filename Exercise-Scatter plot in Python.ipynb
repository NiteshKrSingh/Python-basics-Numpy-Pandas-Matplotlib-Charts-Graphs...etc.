```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#downloading world bank immigration report as on yr 2017
world_migration = pd.read_excel('world_migration_report.xls')
world_migration = world_migration.dropna()
world_migration.head()
```

```python
df_top5 = world_migration.sort_values(['2017'], ascending=False, axis=0).head(5)
df_top5
```

```python
plt.scatter(x='country_name', y='2017', data=df_top5)
plt.scatter(x='country_name', y='1982', data=df_top5)
plt.scatter(x='country_name', y='1992', data=df_top5)
plt.scatter(x='country_name', y='2002', data=df_top5)
plt.scatter(x='country_name', y='2012', data=df_top5)

plt.title('Scatter plot of top five countries')
plt.xlabel('Countries')
plt.ylabel('Number of immigrants in millions')

plt.legend()
plt.show()
```

```python
df_least5 = world_migration.sort_values(['2017'], ascending=False, axis=0).tail(5)
df_least5
```

```python
fig = plt.figure(figsize=(8,5))

plt.scatter(x='country_name', y='2017', data=df_least5)
plt.scatter(x='country_name', y='1982', data=df_least5)
plt.scatter(x='country_name', y='1992', data=df_least5)
plt.scatter(x='country_name', y='2002', data=df_least5)
plt.scatter(x='country_name', y='2012', data=df_least5)

plt.title('Scatter plot of below five countries')
plt.xlabel('Countries')
plt.ylabel('Number of immigrants in millions')

plt.legend()
plt.show()
```

```python
#Downloading united nation immigration report to Canada
canada = pd.read_excel('Canada.xlsx')
canada.head()
```

```python
canada.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage','Unnamed: 43', 'Unnamed: 44', 'Unnamed: 45', 'Unnamed: 46', 
             'Unnamed: 47', 'Unnamed: 48', 'Unnamed: 49', 'Unnamed: 50'], axis=1, inplace=True)

canada.rename(columns={'OdName':'Country', 'AreaName':'Continent','RegName':'Region'}, inplace=True)

canada.columns = list(map(str, canada.columns))
canada.set_index('Country', inplace=True)

# add total column
canada['Total'] = canada.sum(axis=1)
years = list(map(str, range(1980, 2014)))
print('data dimensions:', canada.shape)

canada.head(2)
```

```python
#we can use the sum() method to get the total population per year
df_tot = pd.DataFrame(canada[years].sum(axis=0))

# change the years to type int (useful for regression later on)
df_tot.index = map(int, df_tot.index)

# reset the index to put in back in as a column in the df_tot dataframe
df_tot.reset_index(inplace = True)

# rename columns
df_tot.columns = ['year', 'total']

# view the final dataframe
df_tot.head()
```

```python
df_tot.plot(kind='scatter', x='year', y='total', figsize=(10, 6), color='darkblue')
plt.title('Total Immigration to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.show()
```

```python
#Get the equation of line of best fit
x = df_tot['year']
y = df_tot['total']
fit = np.polyfit(x, y, deg=1)
fit
```

```python
df_tot.plot(kind='scatter', x='year', y='total', figsize=(10, 6), color='darkblue')
plt.title('Total Immigration to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')

# plot line of best fit
plt.plot(x, fit[0] * x + fit[1], color='red') # recall that x is the Years
plt.annotate('y={0:.0f} x + {1:.0f}'.format(fit[0], fit[1]), xy=(2000, 150000))
plt.show()

# print out the line of best fit
'No. Immigrants = {0:.0f} * Year + {1:.0f}'.format(fit[0], fit[1])
```

```python
df_countries = canada.loc[['India', 'China', 'Pakistan'], years].transpose()

 # create df_total by summing across three countries for each year
df_total = pd.DataFrame(df_countries.sum(axis=1))

 # reset index in place
df_total.reset_index(inplace=True)

 # rename columns
df_total.columns = ['year', 'total']

 # change column year from string to int to create scatter plot
df_total['year'] = df_total['year'].astype(int)

 # show resulting dataframe
df_total.head()
```

```python
# generate scatter plot
df_total.plot(kind='scatter', x='year', y='total', figsize=(10, 6), color='darkblue')

 # add title and label to axes
plt.title('Immigration from India, China and Pakistan to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')

 # show plot
plt.show()
```
