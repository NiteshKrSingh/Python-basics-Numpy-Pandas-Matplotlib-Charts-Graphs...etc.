```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
world_population = pd.read_excel('world_population.xls')
world_population.sort_values(by='2020', ascending=False, axis=0, inplace=True)
world_population.set_index('country_name', inplace=True)
world_population.head()
```

```python
world_population.tail()
```

```python
years = list(map(str, range(1981, 2020)))
world_population.fillna(0, inplace = True)
```

```python
world_population.drop(columns = ['1960','1961','1962','1963','1964','1965','1966','1967','1968','1969','1970','1971','1972',
                                 '1973','1974','1975','1976','1977','1978','1979','1980'], axis=1, inplace=True)
world_population.head(2)
```

```python
world_population['2019'].head()
```

```python
world_population['2019'].tail()
```

```python
count, bin_edges = np.histogram(world_population['2019'])

print(count)
print(bin_edges)
```

```python
world_population['2019'].plot(kind='hist', figsize=(8,5), xticks=bin_edges)

plt.title('Histogram of population of all 217 countries in 2019')
plt.ylabel('Number of Countries')
plt.xlabel('Total Population')

plt.show()
```

```python
world_population.loc[['India','China','United States','Russian Federation'], years]
```

```python
world_population.loc[['India','China','United States','Russian Federation'], years].plot.hist()

#It does not look right!
```

```python
df = world_population.loc[['India','China','United States','Russian Federation'], years].transpose()
df.head()
```

```python
df.plot(kind='hist', figsize=(10,6),
       bins=15,
       alpha=0.6,
       xticks=bin_edges,
       color=['coral', 'darkslateblue', 'mediumseagreen'])

plt.title('Histogram of Population of India, China, United States & Russia from 1981 to 2019')
plt.ylabel('Number of Years')
plt.xlabel('Total population in Billion')

plt.show()
```

```python
count, bin_edges = np.histogram(df, 15)
xmin = bin_edges[0] - 10
xmax = bin_edges[-1] + 10

df.plot(kind='hist', figsize=(10,6),
       bins=15,
       xticks=bin_edges,
       color=['coral', 'darkslateblue', 'mediumseagreen'],
       stacked = True,
       xlim = (xmin, xmax)
       )

plt.title('Histogram of Population of India, China, United States & Russia from 1981 to 2019')
plt.ylabel('Number of Years')
plt.xlabel('Total population in Billion')

plt.show()
```
