```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

# Bar chart

```python
world_population = pd.read_excel('world_population.xls')
world_population.head()
```

```python
world_population.drop(columns = ['1960','1961','1962','1963','1964','1965','1966','1967','1968','1969','1970','1971','1972',
                                 '1973','1974','1975','1976','1977','1978','1979','1980','1981','1982','1983','1984',
                                 '1985','1986','1987','1988','1989','1990'], axis=1, inplace=True)
world_population.set_index('country_name', inplace=True)
world_population.head(2)
```

```python
years = list(map(str, range(1991, 2020)))

fig = plt.figure(figsize=(10,3))
world_population.loc['India', years].plot(kind = 'bar')
plt.show()
```

```python
#To check the population of India & China (style-1)
#It is not looking good

years = list(map(str, range(1991, 2020)))

fig = plt.figure(figsize=(10,3))
world_population.loc[['India','China'], years].plot(kind = 'bar')
plt.show()
```

```python
years = ['1991','2011','2020']
countries = ['India','China','United States','Russian Federation','Japan']
df = world_population.groupby('country_name').agg('mean')
df.head()
```

```python
df2 = df.loc[countries, years]
df2.plot(kind='bar')
df2.head()
```

```python
income_gp = world_population.groupby('income_group')['1991','2001','2011','2020'].agg('mean')
income_gp.head()
```

```python
income_gp.plot(kind='bar')
```

```python
#Top five contries with high population

years = ['1991','2001','2011','2020']

inplace = True
world_population.sort_values(by='2020', ascending=False, axis=0, inplace=True)

df_top5 = world_population.head(5)
df_top5 = df_top5[years].transpose()

print(df_top5)

df_top5.index = df_top5.index.map(int)
df_top5.plot(kind='bar', figsize=(14, 8), alpha=1)
plt.title('Top five countries in population')
plt.xlabel('Year')
plt.ylabel('Population in billion')
plt.show()
```
