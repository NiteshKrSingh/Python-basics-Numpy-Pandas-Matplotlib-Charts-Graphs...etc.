```python
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
```

# Working with Pearl Global industries ltd(PGIL) data

```python
pgil = pd.read_excel('pgil.xlsx')
pgil
```

```python
pgil.drop(['DematFlagExist'], axis=1, inplace=True)
pgil.rename(columns = {'Year':'year','Open Price':'open_price','High Price':'high_price','Low Price':'low_price','Close Price':'close_price','No.of Shares':'no_of_shares','No. of Trades':'no_of_trades','Total Turnover in Rs':'turnover_in_rs','Deliverable Quantity':'deliverable_quantity'}, inplace=True)
pgil.head(2)
```

```python
print(plt.style.available)
mpl.style.use(['ggplot'])
```

```python
fig = plt.figure(figsize=(8,3))
x=pgil['year']
y=pgil['close_price']
plt.title('Pearl global industries LTD share price details (in Rs)')
plt.xlabel('Year')
plt.ylabel('Final share closed price')
plt.plot(x, y, alpha=1, color='#a800b1', linestyle='-', linewidth=0.5, marker='.', markeredgecolor='red', 
         markeredgewidth=1, markersize=10, markerfacecolor='yellow', visible=True)
plt.show()
```

```python
fig = plt.figure(figsize=(8,3))
x=pgil['year']
y=pgil['close_price']
plt.title('Pearl global industries LTD share price details (in Rs)')
plt.xlabel('Year')
plt.ylabel('Final share closed price')
plt.plot(x, y,label='Close price of shares')
plt.plot(pgil['year'],pgil['open_price'],label='Open price of shares')
plt.legend()
plt.show()
```

# Working with world_population data

```python
world_population = pd.read_excel('world_population.xls')
world_population.head()
```

```python
world_population.drop(columns = ['1960','1961','1962','1963','1964','1965','1966','1967','1968','1969','1970','1971','1972',
                                 '1973','1974','1975','1976','1977','1978','1979','1980','1981','1982','1983','1984',
                                 '1985','1986','1987','1988','1989','1990'], axis=1, inplace=True)
world_population.set_index('Country Name', inplace=True)
world_population.index.name = None
world_population.head(2)
```

```python
years = list(map(str, range(1991, 2020)))

fig = plt.figure(figsize=(10,3))
world_population.loc['India', years].plot(kind = 'line')
plt.title('Population of India from 1991 to 2020')
plt.xlabel('Year')
plt.ylabel('Total population in billion')

plt.show()
```

```python
#To check the population of India & China (style-1)
#It is not looking good

years = list(map(str, range(1991, 2020)))

fig = plt.figure(figsize=(10,3))
world_population.loc[['India','China'], years].plot(kind = 'line')
plt.title('Population of India & China from 1991 to 2020')
plt.xlabel('Year')
plt.ylabel('Total population in billion')

plt.show()
```

```python
#To check the population of India & China (style-2)

years = list(map(str, range(1991, 2020)))

fig = plt.figure(figsize=(10,5))
world_population.loc['India', years].plot(kind = 'line', label = 'Population of India')
world_population.loc['China', years].plot(kind = 'line', label = 'Population of China')
plt.title('Population of India & China from 1991 to 2020')
plt.xlabel('Year')
plt.ylabel('Total population in billion')
plt.legend()
plt.show()
```

```python
#Top five contries with high population

years = list(map(str, range(1991, 2020)))

inplace = True
world_population.sort_values(by='2020', ascending=False, axis=0, inplace=True)

df_top5 = world_population.head(5)
df_top5 = df_top5[years].transpose()

print(df_top5)

df_top5.index = df_top5.index.map(int)
df_top5.plot(kind='line', figsize=(14, 8))
plt.title('Top five countries in population')
plt.xlabel('Year')
plt.ylabel('Population in billion')
plt.show()
```
